# Imports

```python
# Import the libraries needed for data manipulation, plotting, and regression models.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

# Data Loading

```python
# Store the dataset URL so it can be loaded into pandas.
url = "https://github.com/ageron/data/raw/main/lifesat/lifesat.csv"
```

```python
# Load the life satisfaction dataset from the remote CSV file.
df = pd.read_csv(url)
```

```python
# Display the first few rows to confirm the dataset loaded correctly.
df.head()
```

# Prepare Data

```python
# Separate GDP per capita as the input feature and life satisfaction as the target.
X = df[['GDP per capita (USD)']]
y = df[['Life satisfaction']]
```

```python
# Plot the relationship between GDP per capita and life satisfaction.
plt.scatter(X, y, c='red', alpha=0.65, edgecolors='black')
plt.ylim((4, 10))
plt.grid(True)
plt.xlabel('GDP Per Capita (USD)')
plt.ylabel('Life Satisfaction')
plt.show()
```

# Specify and Train Model

### Model Based Algorithm

```python
# Create a linear regression model for the model-based approach.
model_lin = LinearRegression()
```

```python
# Train the linear regression model on the GDP and life satisfaction data.
model_lin.fit(X, y)
```

```python
# Print the learned slope of the fitted linear regression model.
print(f'Model Slope: {model_lin.coef_[0][0]}')
```

```python
# Print the learned intercept of the fitted linear regression model.
print(f'Model Intercept: {model_lin.intercept_[0]}')
```

### Instance Based Algorithm

```python
# Create a K-nearest neighbors regressor for the instance-based approach.
model_knn = KNeighborsRegressor(n_neighbors=3)
```

```python
# Train the K-nearest neighbors regressor on the same dataset.
model_knn.fit(X, y)
```

# Making Prediction

#### Prediction Using Model Based Algorithm

```python
# Predict life satisfaction for Puerto Rico using the linear regression model.
y_pred = model_lin.predict([[33_442.8]])
```

```python
# Print the linear regression prediction for Puerto Rico.
print(f'Life Satisfaction Score for Puerto Rico (Linear Regression): {y_pred[0][0]}')
```

#### Prediction Using Instance Based Algorithm

```python
# Predict life satisfaction for Puerto Rico using the KNN regressor.
y_pred = model_knn.predict([[33_442.8]])
```

```python
# Print the KNN prediction for Puerto Rico.
print(f'Life Satisfaction Score for Puerto Rico (KNN): {y_pred[0][0]}')
```

# Regression Plot (Linear Regression)

```python
# Plot the fitted linear regression line alongside the original data points.
plt.scatter(X, y, c='red', alpha=0.65, edgecolors='black')
plt.axline(xy1=(0, model_lin.intercept_[0]), slope=model_lin.coef_[0][0], c='green', ls='-.')
plt.ylim((4, 10))
plt.grid(True)
plt.xlabel('GDP Per Capita (USD)')
plt.ylabel('Life Satisfaction')
plt.show()
```
